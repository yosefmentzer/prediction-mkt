```python
footy_team_names_fuzzy_for_manual_check_names = footy_team_names_fuzzy_for_manual_check[footy_team_names_fuzzy_for_manual_check['row_type']=='bf_team_names'].copy()
```

```python
merged = footy_team_names_fuzzy_for_manual_check_names.merge(previous_df_bf_teamnames[['team_name','country_iso','names']],
                                                    how='left',
                                                   left_on=['team_name','country_code'],
                                                   right_on=['team_name','country_iso'])
merged
```

```python
bf_team_cols = [f'bf_team_{i}' for i in range(0,13)]
bf_team_cols
```

```python
merged['candidates'] = merged[bf_team_cols].apply(lambda x: [a for a in x], axis=1)
```

```python
zero_ones = merged.apply(lambda x: ['1' if a in x['names'] else '0' for a in x['candidates']], axis=1).apply(pd.Series)
```

```python
zero_ones.columns = bf_team_cols
```

```python
zero_ones
```

```python
merged_zero_ones = merged.copy()
```

```python
for i in range(len(merged_zero_ones)):
    for j in range(0,13):
        merged_zero_ones[f'bf_team_{j}'].iloc[i] = zero_ones[f'bf_team_{j}'].iloc[i]
```

```python
merged_zero_ones
```

```python
merged.index = range(0, 2*len(merged), 2)
```

```python
merged_zero_ones.index = range(1, 2*len(merged_zero_ones)+1, 2)
```

```python
# write row type before concatenating DFs
merged['row_type'] = 'bf_team_names'
merged_zero_ones['row_type'] = 'name_matching'
```

```python
footy_team_names_fuzzy_for_manual_check_with_previous = pd.concat([merged, merged_zero_ones]).drop(
    columns=['country_iso','names','candidates']).sort_index()
footy_team_names_fuzzy_for_manual_check_with_previous
```

```python
footy_team_names_fuzzy_for_manual_check_with_previous.to_csv(os.path.join(INTERIM_FOLDERPATH, 
                                                            'footy_team_names_fuzzy_for_manual_check_with_previous.csv'), 
                                               index=False)
```
