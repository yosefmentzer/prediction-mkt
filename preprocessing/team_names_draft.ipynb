```python
FOOTY_TEAM_NAMES_SCORES_EXPLODED_PATH = '../data/interim/footy_team_names_scores_exploded.csv'
df = pd.read_csv(FOOTY_TEAM_NAMES_SCORES_EXPLODED_PATH)
```

```python
BF_TEAM_NAMES_PATH = '../data/interim/bf_team_names.csv'
bf_team_names = pd.read_csv(BF_TEAM_NAMES_PATH)
list(bf_team_names['team_name'])[:20]
```

```python
for i, (country_code, team_name) in enumerate(zip(df_matches['country_code'], df_matches['team_name'])):
    if manual_matching[(country_code, team_name)]:
        fr_matches = manual_matching[(country_code, team_name)]  # list
        for column, bf_team in zip(df.iloc[i][cols].index, df.iloc[i][cols].values):
            if bf_team in fr_matches:
                df_matches.loc[i, column] = 1
        manual_insertion_list = []
        for fr_match in fr_matches:
            if (fr_match not in list(df.iloc[i][cols].values)) and (fr_match in list(bf_team_names['team_name'])):
                manual_insertion_list.append(fr_match)
        if len(manual_insertion_list) == 0:
            df_matches.loc[i, 'manual_insertion'] = 0
            df.loc[i, 'manual_insertion'] = np.nan
        elif len(manual_insertion_list) == 1:
            df_matches.loc[i, 'manual_insertion'] = 1
            df.loc[i, 'manual_insertion'] = manual_insertion_list[0]
        else:
            df_matches.loc[i, 'manual_insertion'] = 1
            df.loc[i, 'manual_insertion'] = str(manual_insertion_list)
df_matches['manual_insertion'] = df_matches['manual_insertion'].astype(int)
```
